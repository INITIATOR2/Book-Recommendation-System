# Merging DataFrames

Merging all the above datasets(Books,Ratings,Users).

There are three different types of merges available in Pandas given below:-

Inner Merge / Inner join .
Left Merge / Left join
Right Merge / Right join
Outer Merge / Full outer join
The merge type to use is specified using the “how” parameter in the merge command, taking values “left”, “right”, “inner” (default), or “outer”.

```python
outer_join = pd.merge(ratings, books, on ='ISBN',how= 'outer',indicator=True)
outer_join
```

```python
outer_join._merge.value_counts()
```

```python
# #unmmatched data betwen Ratings and Book dataset
mismatch = outer_join[outer_join._merge.isin(['left_only','right_only'])]
mismatch
```

```python
mismatch.isnull().sum()
```

```python
# Heatmap for null values
plt.figure(figsize=(20,10))
plt.title("Heatmap for null values in merged dataset",fontsize = 15)
sns.heatmap(mismatch.isnull(), cbar=True,cmap="YlGnBu")
```

```python
#Inner join between ratings and book dataset
inner_join = pd.merge(ratings , books, on ='ISBN',how= 'inner')
inner_join
ratings_with_name=inner_join
```

```python
#Inner join between inner_join and Users dataset
inner_join1 = pd.merge(inner_join, users, on ='User-ID')
inner_join1
```

# Collaborative Filtering Based Recommender System

```python
ratings_with_name   # inner join between rating and book dataset
```

filtering user on the basis of vote 
sellected only those user who can vote more than 200

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] >200
active_users= x[x].index
active_users
```

```python
# selecting those users which vote more than 200 books
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(active_users)]
filtered_rating
```

```python
#filtering based on book title
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50 # top 50 reated book
famous_books = y[y].index
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values = 'Book-Rating')  # pivot table on final rating for Querying large amount of data in many user-Friendly Ways
```

```python
pt.fillna(0,inplace=True)
```

```python
pt  # this table includes those people who vote more than 200 and those books whose rating is more than 200
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
simalarity_Scores = cosine_similarity(pt)   # for calculating euclidian Distance from 1 point to all porint
```
